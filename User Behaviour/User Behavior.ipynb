Анализ для крупного интернет-магазина.

Необходимо приоритизировать гипотезы, запустить A/B-тест и проанализировать результаты. 

По результатам теста необходимо принять решение. Варианты решений:

# Описание данных

1. Данные для первой части

``Hypothesis`` — краткое описание гипотезы;  
``Reach`` — охват пользователей по 10-балльной шкале;  
``Impact`` — влияние на пользователей по 10-балльной шкале;  
``Confidence`` — уверенность в гипотезе по 10-балльной шкале;  
``Efforts`` — затраты ресурсов на проверку гипотезы по 10-балльной шкале. Чем больше значение Efforts, тем дороже проверка гипотезы.

2. Данные для второй части

Файл ``orders``:  

``transactionId`` — идентификатор заказа;  
``visitorId`` — идентификатор пользователя, совершившего заказ;  
``date`` — дата, когда был совершён заказ;  
``revenue`` — выручка заказа;  
``group`` — группа A/B-теста, в которую попал заказ.

Файл ``visitors``:

``date`` — дата;  
``group`` — группа A/B-теста;  
``visitors`` — количество пользователей в указанную дату в указанной группе A/B-теста

## Приоритизация гипотез.

### Приоритизации гипотез. Фреймворк ICE

```python
pd.set_option('display.max_colwidth', None)
hypothesis['ICE'] = round(hypothesis['Impact'] * hypothesis['Confidence'] / hypothesis['Efforts'], 2)
display(hypothesis.sort_values(by = 'ICE', ascending = False))
```

#### Вывод

Гипотезы ``Запустить акцию, дающую скидку на товар в день рождения``, ``Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей	`` и ``Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок`` имеют самые высокие показатели ICE, что означает, что они являются наиболее перспективными и должны быть рассмотрены в первую очередь при определении приоритетов тестирования.

Все эти 3 гипотезы имеют наивысший показатель ICE благодаря высоким оценкам в Impact и Confidence. При этом Efforts оценена на уровне среднего уровня. Это может означать, что реализация этой гипотезы потребует некоторых усилий, но она может принести значительное улучшение метрик.

### Приоритизации гипотез. Фреймворк RICE

```python
hypothesis['RICE'] = hypothesis['Reach'] * hypothesis['Impact'] * hypothesis['Confidence'] / hypothesis['Efforts']
display(hypothesis.sort_values(by = 'RICE', ascending = False))
```

#### Вывод

Гипотезы ``Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок	``, ``Добавить блоки рекомендаций товаров на сайт интернет магазина, чтобы повысить конверсию и средний чек заказа`` и ``Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей	`` имеют наибольшие показатели RICE, что означает, что они являются наиболее перспективными и должны быть рассмотрены в первую очередь при определении приоритетов тестирования.

Гипотеза ``Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок	`` поднялась на первое место и имеет наивысший показатель RICE благодаря высокой оценке Reach. Это может означать, что изменения, связанные с этой гипотезой, затронут большое число пользователей и могут привести к значительному улучшению метрик бизнеса.

Гипотеза ``Добавить блоки рекомендаций товаров на сайт интернет магазина, чтобы повысить конверсию и средний чек заказа`` также сильно поднялась после оценки RICE благодаря высокой оценке Reach. При этом Impact имеет относительно низкую оценку. Это может означать, что ее влияние на метрики бизнеса может быть значительным.

### Вывод

При использовании показателя ICE для приоритизации гипотез наиболее перспективными являются гипотезы ``Запустить акцию, дающую скидку на товар в день рождения`` и ``Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей	``. Эти гипотезы имеют высокие показатели Impact и Confidence, относительно средний Efforts, но низкий Reach

Однако, при использовании показателя RICE, гипотезы ``Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок	`` и ``Добавить блоки рекомендаций товаров на сайт интернет магазина, чтобы повысить конверсию и средний чек заказа`` оказались наиболее перспективными. Это объясняется тем, что они имеют высокие оценки Reach, а также уверенность в том, что изменения повлияют на метрики бизнеса. Гипотеза ``Запустить акцию, дающую скидку на товар в день рождения``, которая была высоко оценена по показателю ICE, оказались менее перспективной по показателю RICE, что связано с ее низкой оценкой Reach.

Таким образом, при выборе гипотез для тестирования, я бы рекомендовал использовать показатель RICE, так как он учитывает не только потенциальное влияние на метрики бизнеса, но и охват пользователей, а значит гипотезы ``Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок	`` и ``Добавить блоки рекомендаций товаров на сайт интернет магазина, чтобы повысить конверсию и средний чек заказа`` являются наиболее перспективными

### Построение графика кумулятивной выручки по группам

```python
# делаем заказы и выручку по дням
ordersADaily = (
    orders[orders['group'] == 'A'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']

ordersBDaily = (
    orders[orders['group'] == 'B'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue': 'sum'})
)
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']
```

```python
# делаем заказы и выручку куммулятивно
ordersACummulative = ordersADaily.apply(
    lambda x: ordersADaily[ordersADaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateA': 'sum', 'revenuePerDateA': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersACummulative.columns = [
    'date',
    'ordersCummulativeA',
    'revenueCummulativeA',
]


ordersBCummulative = ordersBDaily.apply(
    lambda x: ordersBDaily[ordersBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateB': 'sum', 'revenuePerDateB': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersBCummulative.columns = [
    'date',
    'ordersCummulativeB',
    'revenueCummulativeB',
]
```

```python
# строим график
plt.figure(figsize=(15,6))
plt.ticklabel_format(style='plain')
plt.plot(ordersACummulative['date'], ordersACummulative['revenueCummulativeA'], label = 'A')
plt.plot(ordersBCummulative['date'], ordersBCummulative['revenueCummulativeB'], label = 'B')
plt.legend()
plt.xticks(ticks = ordersACummulative['date'], fontsize = 11, rotation=45)
plt.title('График кумулятивной выручки по группам', fontsize=15, fontweight="bold")
plt.xlabel('Дата', fontsize=13)
plt.ylabel('Кумулятивная выручка', fontsize=13)
plt.grid()
```

На графике кумулятивной выручки по группам видно, что начиная с первой недели тестирования группа В опережает группу А по суммарной выручке. В течение первых двух недель тестирования графики выручки для обеих групп имели схожий характер, но начиная с 3-й недели группа В пошла вверх более быстрыми темпами и увеличила свой отрыв. ### Расчет статистической значимости различий в среднем количестве заказов на посетителя между группами по «сырым» данным

Для расчета статистической значимости примем нулевую и альтернтивную гипотезы:  
Н0: Статистически значимых различий в среднем количестве заказов на посетителя между группами нет.  
Н1: Существуют статистически значимые различия в среднем количестве заказов на посетителя между группами.

```python
# сгруппируем по кол-ву заказов на пользователя в группе А
ordersByUsersA = (
    orders[orders['group'] == 'A']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': 'nunique'})
)
ordersByUsersA.columns = ['visitorId', 'transactions']

# сгруппируем по кол-ву заказов на пользователя в группе В
ordersByUsersB = (
    orders[orders['group'] == 'B']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': 'nunique'})
)
ordersByUsersB.columns = ['visitorId', 'transactions']
```

```python
# проведем тест уитни-манна
sampleA = pd.concat([ordersByUsersA['transactions'],
                     pd.Series(0, 
                               index = np.arange(visitorsA['visitors'].sum() - len(ordersByUsersA['transactions'])), 
                               name = 'transactions')], axis = 0)

sampleB = pd.concat([ordersByUsersB['transactions'],
                     pd.Series(0, 
                               index = np.arange(visitorsB['visitors'].sum() - len(ordersByUsersB['transactions'])), 
                               name = 'transactions')], axis = 0)
```

```python
print("Значение p-value: {0:.3f}".format(stats.mannwhitneyu(sampleA, sampleB)[1]))
print("Относительный прирост группы В к группе А: {0:.3f}".format(sampleB.mean() / sampleA.mean() - 1))
```

#### Вывод

Относительный прирост количества заказов в группе B по сравнению с группой A составляет 16%.  
P-value меньше уровня критической значимости (0.05), нулевую гипотезу можно отвергнуть.    
Это означает, что группа B показывает лучшие результаты по количеству заказов на пользователя по сравнению с группой A, и это различие является статистически значимым.

### Расчет статистической значимости различий в среднем чеке заказа между группами по «сырым» данным

Для расчета статистической значимости примем нулевую и альтернтивную гипотезы:  
Н0: Статистически значимых различий в среднем чеке заказа между группами нет.  
Н1: Существуют статистически значимые различия среднем чеке заказа между группами.

### Расчет статистической значимости различий в среднем количестве заказов на посетителя между группами по «очищенным» данным

Гипотезы аналогичны расчетам по "сырым" данным:  
Н0: Статистически значимых различий в среднем количестве заказов на посетителя между группами нет.  
Н1: Существуют статистически значимые различия в среднем количестве заказов на посетителя между группами.

```python
# приступим к очистке данных
# отсеим пользователей с большим кол-вом заказов

usersWithManyOrders = pd.concat(
    [
        ordersByUsersA[ordersByUsersA['transactions'] > abnormal_orders]['visitorId'],
        ordersByUsersB[ordersByUsersB['transactions'] > abnormal_orders]['visitorId'],
    ],
    axis=0,
)
```

```python
# отсеим пользователей с дорогими заказами

usersWithExpensiveOrders = orders[orders['revenue'] > abnormal_revenue]['visitorId']
abnormalUsers = (
    pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis=0)
    .drop_duplicates()
    .sort_values()
)
```

```python
# объединим пользоваталей с многими заказами с пользователями с дорогими заказами
abnormalUsers = (
    pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis=0)
    .drop_duplicates()
    .sort_values()
)

print('Количество аномальных пользователей:', abnormalUsers.shape[0]) 
```

```python
# сделаем отфильтрованны сэмлы, где не будет аномальных пользователей

sampleAFiltered = pd.concat([ordersByUsersA[np.logical_not(
            ordersByUsersA['visitorId'].isin(abnormalUsers))]['transactions'],
                     pd.Series(0, 
                               index = np.arange(visitorsA['visitors'].sum() - len(ordersByUsersA['transactions'])), 
                               name = 'transactions')], axis = 0)

sampleBFiltered = pd.concat([ordersByUsersB[np.logical_not(
            ordersByUsersB['visitorId'].isin(abnormalUsers))]['transactions'],
                     pd.Series(0, 
                               index = np.arange(visitorsB['visitors'].sum() - len(ordersByUsersB['transactions'])), 
                               name = 'transactions')], axis = 0)
```

```python
print("Значение p-value: {0:.3f}".format(stats.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]))

print("Относительный прирост группы В к группе А: {0:.3f}".format(sampleBFiltered.mean() / sampleAFiltered.mean() - 1))
```

#### Вывод

P-value меньше уровня критической значимости (0.05), нулевую гипотезу по-прежнему можно отвергнуть.  
Исключение аномальных пользователей повлияло на результат следующим образом:
относительный прирост количества заказов в группе B по сравнению с группой A вырос с 16% до 19.7%.

### Расчет статистической значимости различий в среднем чеке заказа между группами по «очищенным» данным
